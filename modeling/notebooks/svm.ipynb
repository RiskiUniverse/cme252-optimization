```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from cvxpy import *
```

```python
dim = 2
r = 1
std = .1
N = 50

c = np.random.randn(dim)
c = r*c/np.linalg.norm(c)

pos = std*np.random.randn(dim, N) + c[:, np.newaxis]
neg = std*np.random.randn(dim, N) - c[:, np.newaxis]
```

```python
plt.scatter(pos[0], pos[1], c=.1*np.ones(N), alpha=.5)

plt.scatter(neg[0], neg[1], c=10*np.ones(N), alpha=.5)
```
